```python
import json

path = './output.json'

with open(path, 'r') as f:
    data = json.load(f)

sample = []
sources = set()

while len(sources) != 3:
    for i in data:
        if i['source'] not in sources:
            sources.add(i['source'])
            sample.append(i)

        if len(sources) == 3:
            break

sample
```

```python
# print folder structure of '../../backend'

import os

def print_folder_structure(path):
    for root, dirs, files in os.walk(path):
        if 'node_modules' in dirs:
            dirs.remove('node_modules')
        level = root.replace(path, '').count(os.sep)
        indent = ' ' * 4 * (level)
        print(f'{indent}{os.path.basename(root)}/')
        sub_indent = ' ' * 4 * (level + 1)
        for f in files:
            print(f'{sub_indent}{f}')

print_folder_structure('../../backend')
```
